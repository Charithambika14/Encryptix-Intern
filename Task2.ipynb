**Movie Rating Prediction using Python**

- We will predicte the rating of movie with help of which genre, director, actors.

- Using linear Regression, Y = X*b + e
  *   Y = rating
  *   X = matrix which contains name, year, genre, director, actor1, actor2, actor3.
  *   b = unknown parameter which will be estimated by model training
  *   e = error term

Now all columns have changed to numeric values

```python
features=data.drop("Rating",axis=1)
target=data["Rating"]
```

```python
values=mutual_info_regression(features,target)
mutual_info=pd.Series(values)
mutual_info.index=features.columns
df=pd.DataFrame({"Feature":mutual_info.index,"Values":mutual_info})
df.sort_values(ascending=True,by="Values",inplace=True)
df["Values"]=df["Values"].round(3)
bar2=sns.barplot(x="Feature",y="Values",data=df)
bar2.bar_label(bar2.containers[0])
plt.title("MUTUAL INFORMATION GAIN")
plt.show()
```

Duration column is not signficant in predicting rating of movie so we drop it

```python
data.drop("Duration",axis=1,inplace=True) ## correlation(duration,rating) = 0
```

```python
features=data.drop("Rating",axis=1)
```

Splitting the dataset into 95% training dataset and 5% testing dataset

```python
x_train,x_test,y_train,y_test=train_test_split(features,target,test_size=0.05,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

Model training

```python
model=LinearRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
```

Performance evaluation using Mean squared error and R squared error

```python
mse=mean_squared_error(pred,y_test)
round(mse,2)
```

```python
score=r2_score(pred,y_test)
round(score,2)
```

```python
sns.regplot(x=y_test,y=pred,line_kws=dict(color="r"),scatter_kws=dict(color="green"))
plt.xlabel("Actual Rating")
plt.ylabel("Predicted Rating")
plt.title("Actual vs Predicted Rating")
plt.show()
```

```python
residual=y_test-pred
plt.scatter(x=pred,y=residual,c="green")
plt.xlabel("Predicted Values")
plt.ylabel("Error")
plt.title("Residual Error")
plt.show()
```
